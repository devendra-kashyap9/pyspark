### Spark Session

### ⛔️⚠️ Run this on priority 

```python
import os
import sys
import shutil

# 1. Update the paths to point directly to Java 17
os.environ["JAVA_HOME"] = "/opt/homebrew/opt/openjdk@17"
os.environ["PATH"] = "/opt/homebrew/opt/openjdk@17/bin:" + os.environ.get("PATH", "")
os.environ["PYSPARK_PYTHON"] = sys.executable
os.environ["PYSPARK_DRIVER_PYTHON"] = sys.executable

print(f"Target Java: {shutil.which('java')}")

# 2. Initialize Spark
from pyspark.sql import SparkSession

try:
    spark = SparkSession.builder \
        .master("local[1]") \
        .appName("Practice") \
        .config("spark.driver.bindAddress", "127.0.0.1") \
        .config("spark.driver.host", "127.0.0.1") \
        .config("spark.ui.port", "4050") \
        .getOrCreate()
    print("🎉 FINALLY! Spark Session successfully created with Java 17!")
except Exception as e:
    print(f"❌ Initialization issue: {e}")
```
